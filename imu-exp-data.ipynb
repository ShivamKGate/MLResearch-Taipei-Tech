```python
#THE AI FUN

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models

imu = pd.read_csv('/kaggle/input/imudataset/IMU_EXP.csv')
imu_melted = imu.melt(id_vars='Time', value_vars=['rate2_v', 'rate1_v', 'rate3_v'], var_name='Rate', value_name='Value')


#THE MAGIC

#normalizing
for rate in ['rate2_v', 'rate1_v', 'rate3_v']:
    imu[rate] = (imu[rate] - imu[rate].mean()) / imu[rate].std()

train_data = imu[['rate2_v', 'rate1_v', 'rate3_v']].values

#VAE start
latent_dim = 2

class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon
    
# encoder
encoder_inputs = layers.Input(shape=(3,))
x = layers.Dense(64, activation='relu')(encoder_inputs)
x = layers.Dense(32, activation='relu')(x)
z_mean = layers.Dense(latent_dim)(x)
z_log_var = layers.Dense(latent_dim)(x)
z = Sampling()([z_mean, z_log_var])
encoder = models.Model(encoder_inputs, [z_mean, z_log_var, z], name='encoder')

# decoder
latent_inputs = layers.Input(shape=(latent_dim,))
x = layers.Dense(32, activation='relu')(latent_inputs)
x = layers.Dense(64, activation='relu')(x)
decoder_outputs = layers.Dense(3)(x)
decoder = models.Model(latent_inputs, decoder_outputs, name='decoder')

# VAE
outputs = decoder(encoder(encoder_inputs)[2])
vae = models.Model(encoder_inputs, outputs, name='vae')

optimizer = tf.keras.optimizers.Adam()

epochs = 100
batch_size = 32
for epoch in range(epochs):
    for i in range(0, len(train_data), batch_size):
        batch = train_data[i:i + batch_size]
    print(f'Epoch {epoch + 1}')

# denoising
denoised_data = vae.predict(train_data)

denoised_df = pd.DataFrame(denoised_data, columns=['rate2_v', 'rate1_v', 'rate3_v'])
denoised_df['Time'] = imu['Time'].values

imu_improved_melted = denoised_df.melt(id_vars='Time', value_vars=['rate2_v', 'rate1_v', 'rate3_v'], var_name='Rate', value_name='Value')    

sns.set(style="whitegrid")

#original data
plt.figure(figsize=(20, 8))
sns.lineplot(data=imu_melted, x='Time', y='Value', hue='Rate', alpha=0.3)
plt.title('Original Data')
plt.show()

#improved data
plt.figure(figsize=(20, 8))
sns.lineplot(data=imu_improved_melted, x='Time', y='Value', hue='Rate')
plt.title('Improved Data')
plt.show()
```
